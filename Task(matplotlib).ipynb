```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
df=pd.read_csv('/content/drive/MyDrive/tip(in).csv')
```

```python
print("(Rows,Columns) : ",df.shape)
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
plt.plot(df['total_bill'],df['time'])
```

```python
plt.bar(df.sex,df.total_bill)
plt.xlabel("Sex")
plt.ylabel("Total Bill")
plt.title("Total Bill vs Sex")
plt.show()
```

```python
plt.hist(df.total_bill,bins=10,color='orange',edgecolor='black')
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.title("Histogram of Total Bill")
plt.show()
```

```python
plt.scatter(df.total_bill,df.tip)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Total Bill vs Tip")
plt.show()
```

```python
plt.boxplot(df.total_bill)
plt.xlabel("Total Bill")
plt.title("Boxplot of Total Bill")
plt.show()
```

```python
gby=df.groupby('day')['total_bill'].sum()
plt.pie(gby,labels=gby.index,autopct='%1.1f%%',startangle=90)
plt.title("Total_bill Per Day")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.bar(df['size'],df['total_bill'])
plt.xlabel("Size")
plt.ylabel("Total Bill")
plt.title("Size vs Total Bill")
plt.subplot(1,2,2)
plt.scatter(df.smoker,df.total_bill)
plt.xlabel("Smoker")
plt.ylabel("Total Bill")
plt.title("Smoker vs Total Bill")
plt.show()
```
