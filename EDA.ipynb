```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
print(df.shape)
print(df.info())
print(df.describe())
```

```python
missing=({
    'missing_count':df.isnull().sum(),
'missing_percentage':(df.isnull().sum()/len(df)*100)
})
missing_df=pd.DataFrame(missing)
missing_df.sort_values(by='missing_percentage',ascending=False)
```

```python
df['age']=df['age'].fillna(df['age'].mean())
df['embarked']=df['embarked'].fillna(df['embarked'].mode()[0])
df.drop(columns=['deck'],inplace=True)
print("missing values after cleaning : \n",df.isnull().sum().sum())
```

```python
missing=({
    'missing_count':df.isnull().sum(),
'missing_percentage':(df.isnull().sum()/len(df)*100)
})
missing_df=pd.DataFrame(missing)
missing_df.sort_values(by='missing_percentage',ascending=False)
```

```python
print(df.duplicated().sum())
df['class']=df['class'].astype('category')
df['sex']=df['sex'].astype('category')
df['embarked']=df['embarked'].astype('category')
df.info()
```

```python
num_cols=df.select_dtypes(include=["int64","float64"]).columns
df[num_cols].describe()
```

```python
sns.histplot(df['age'],bins=30,kde=True,color='skyblue')
plt.title('Age Distribution of passengers')
plt.show()
```

```python
sns.boxplot(x=df['fare'],color='orange')
plt.title('Fare Distribution with outliers')
plt.show()
```

```python
cat_clos=df.select_dtypes('category').columns
for col in cat_clos:
  sns.catplot(data=df,x=col)
  plt.title(f'Count Plot for {col}')
  plt.show()
```

```python
sns.boxplot(x='class',y='fare',data=df)
plt.title('Fare Distribution by Class')
plt.show()
```

```python
sns.boxplot(x='sex',y='age',data=df)
plt.title('Age Distribution by Sex')
plt.show()
```

```python
sns.boxplot(x='survived',hue='sex',data=df)
plt.title('Survival count by gender')
plt.show()
```

```python
sns.boxplot(x='survived',hue='class',data=df)
plt.title('Survival count by class')
plt.show()
```

```python
corr=df[['age','fare','survived']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
sns.pairplot(df[['age','fare','survived']],hue='survived')
plt.show()
```

```python
q1=df['fare'].quantile(0.25)
q3=df['fare'].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
outliers=df[(df['fare']<lower_bound) | (df['fare']>upper_bound)]
print("Number of outliers in fare : ",len(outliers))
df['fare_capped']=df['fare'].clip(lower=lower_bound,upper=upper_bound)
sns.boxplot(x=df['fare_capped'])
plt.title('Fare Distribution with outliers')
plt.show()
```
