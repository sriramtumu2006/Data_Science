```python
import matplotlib.pyplot as plt
months=['jan','feb','march','april','may']
sales=[300,450,80,400,550]
plt.plot(months,sales,marker='o',linestyle='--')
plt.title('monthly sales')
plt.xlabel('months')
plt.ylabel('sales(in $)')
plt.grid('True')
plt.show()
```

```python
products=['TV','Laptop','Mobile','Tablet']
revenue=[35000,50000,20000,25000]
plt.bar(products,revenue,color='purple')
plt.title('Revenue by product')
plt.xlabel('products')
plt.ylabel('revenue(in $)')
plt.show()
```

```python
age=[25,35,40,44,52,58,65]
salary=[20000,29500,45000,50000,60000,69000,80000]
plt.scatter(age,salary,color='red')
plt.title('Age VS Salary')
plt.xlabel('age')
plt.ylabel('salary(in $)')
plt.show()
```

```python
brands=['Apple','Samsung','OnePlus','Vivo','Others']
market_share=[30,28,20,10,12]
colors=['gold','lightblue','lightgreen','pink','grey']
plt.pie(market_share,labels=brands,colors=colors,autopct='%.1f%%',startangle=90)
plt.title('Market Share of Mobile Brands')
plt.show()
```

```python
import numpy as np
ages=np.random.randint(18,60,100)
plt.hist(ages,bins=8,color='violet',edgecolor='black')
plt.title('Age Distribution of Employees')
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y1=[10,40,20,60,50,30]
y2=[5,15,25,35,45,55]
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y1,'r--')
plt.title('Line Plot')
plt.subplot(1,2,2)
plt.bar(x,y2)
plt.title('Bar Plot')
plt.show()
```

```python
plt.style.use('ggplot')
x=[1,2,3,4,5,6]
sales=[10,40,20,60,50,30]
profit=[5,25,35,45,55,15]
plt.figure(figsize=(10,5))
plt.plot(x,sales,label='sales',marker='o')
plt.plot(x,profit,label='profit',marker='s')
plt.title('Sales VS Profit Trend')
plt.xlabel("Quarter")
plt.ylabel("Amount($)")
plt.legend()
plt.savefig('sales_profit.png')
plt.show()
```

```python
dn=pd.read_csv('/content/drive/MyDrive/data.csv')
print(dn.isnull().sum())
print("(Rows,Columns) :",dn.shape)
dn.drop_duplicates(inplace=True)
print("After removing duplicates (Rows,Columns) :",dn.shape)
dn['Date']=pd.to_datetime(dn['Date'])
dn['Date']=dn['Date'].sort_values(ascending=True).copy()
print(dn[dn['Date'].isnull()])
```

```python
dn['Calories']=dn['Calories'].fillna(dn['Calories'].mean())
print(dn.isnull().sum())
print(dn)
```

```python
dn.dropna(inplace=True)
x=np.array(dn['Date'])
y=np.array(dn['Calories'])
plt.figure(figsize=(20,5))
plt.plot(x,y)
plt.title('Calories Over Time')
plt.xlabel('Date')
plt.ylabel('Calories')
plt.savefig('calories_over_time.png')
plt.show()
```

```python
dn.groupby('Duration')['Pulse'].mean().plot(kind='bar',figsize=(10,5))
plt.title('Average Pulse Over Duration')
plt.xlabel('Duration')
plt.ylabel('Average Pulse')
plt.savefig('Average Pulse Over Duration.png')
plt.show()
```

```python
from scipy.stats import pearsonr
plt.scatter(dn['Duration'],dn['Calories'],color='red')
plt.title('Duration VS Calories')
plt.xlabel('Duration')
plt.ylabel('Calories')
correlation,_=pearsonr(dn['Duration'],dn['Calories'])
print('Correlation:',correlation)
m,b=np.polyfit(dn['Duration'],dn['Calories'],1)
plt.plot(dn['Duration'],m*dn['Duration'] +b)
plt.savefig('Scatter_duration_vs_cal.png')
plt.show()
```

```python
ranges=['<=120','121-140','141-160','>160']
r=[
    (dn['Maxpulse']<=120).sum(),
    ((dn['Maxpulse']>120) & (dn['Maxpulse']<=140)).sum(),
    ((dn['Maxpulse']>140) & (dn['Maxpulse']<=160)).sum(),
    (dn['Maxpulse']>160).sum()
]
colors=['gold','cyan','violet','red']
plt.pie(r,labels=ranges,colors=colors,autopct='%.1f%%',startangle=90)
plt.title('Max Pulse Ranges')
plt.savefig('pie_maxpulse_ranges.png')
plt.show()
```

```python
plt.hist(dn['Calories'],bins=10,color='green',edgecolor='black')
plt.title('Calories Distribution')
plt.xlabel('Calories')
plt.ylabel('Count')
plt.savefig('hist_calories.png')
plt.show()
```

```python
plt.figure(figsize=(22,6))
plt.subplot(1,2,1)
plt.plot(dn['Date'],dn['Pulse'],color='red')
plt.title('Pulse over Date')
plt.xlabel('Date')
plt.ylabel('Pulse')
plt.subplot(1,2,2)
plt.scatter(dn['Duration'],dn['Maxpulse'],color='blue')
plt.title('Duration VS Maxpulse')
plt.xlabel('Duration')
plt.ylabel('Max Pulse')
plt.savefig('subplot_pulse_maxpulse.png')
plt.show()
```
