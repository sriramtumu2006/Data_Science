```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv('/content/drive/MyDrive/amazon_products_dataset.csv')
```

```python
df.head()
```

```python
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.shape)
```

```python
df.isnull().sum()
```

```python
df['actual_price']=df['actual_price'].fillna(df['actual_price'].mean())
df['discount_price']=df['discount_price'].fillna(df['discount_price'].mean())
df['rating']=df['rating'].fillna(df['rating'].mean())
df['rating_count']=df['rating_count'].fillna(df['rating_count'].mean())
```

```python
df.isnull().sum()
```

```python
df.fillna('unknown',inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df.reset_index(drop=True,inplace=True)
df['rating_count']=df['rating_count'].astype(int)
```

```python
df.sort_values(by='actual_price',ascending=False).head()
```

```python
df.groupby('brand')['category'].count().idxmax()
```

```python
df['dis_per']=(df['actual_price']-df['discount_price'])/df['actual_price']*100
df.groupby('brand')['dis_per'].mean()
```

```python
df[(df['rating']>=4.5) & (df['rating_count']>=2000)]
```

```python
df.groupby('category')['actual_price'].mean()
```

```python
import numpy as np
import math
df['discount_percent'] = ((df['actual_price'] - df['discount_price']) / df['actual_price']) * 100
df['price_category'] = pd.cut(
    df['actual_price'],
    bins=[-np.inf, 500, 2000, np.inf],
    labels=['Budget', 'Midrange', 'Premium']
)
df['popularity_score'] = df['rating'] * np.log(df['rating_count'] + 1)
```

```python
top10_popular = df.sort_values('popularity_score', ascending=False).head(10)
```

```python
premium_high_discount = df[
    (df['price_category'] == 'Premium') &
    (df['discount_percent'] > 40)
]
```

```python
bluetooth_items = df[df['title'].str.contains("Bluetooth", case=False, na=False)]
```

```python
popularity_threshold = df['popularity_score'].quantile(0.70)
filtered_products = df[
    (df['rating'] >= 4.0) &
    (df['discount_percent'].between(20, 50)) &
    (df['popularity_score'] >= popularity_threshold)
]
print(filtered_products)
```

```python
avg_discount_per_brand = df.groupby('brand')['discount_percent'].mean()
print(avg_discount_per_brand)
```

```python
avg_popularity_by_category = df.groupby('price_category')['popularity_score'].mean()
print(avg_popularity_by_category)
```

```python
category_best_discount = (
    df.groupby('price_category')['discount_percent']
    .mean()
    .sort_values(ascending=False)
)
print(category_best_discount)
```

```python
df['revenue_potential'] = df['actual_price'] * df['rating_count']
revenue_per_brand = df.groupby('brand')['revenue_potential'].sum()
print(revenue_per_brand)
```

```python
df.to_csv('/content/drive/My Drive/cleaned_amazon_products_dataset.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='actual_price', y='rating', data=df)
plt.title('Actual Price vs. Rating')
plt.xlabel('Actual Price')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['actual_price'], kde=True)
plt.title("Price Distribution (Histogram + KDE)")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
avg_rating = df.groupby('category')['rating'].mean().reset_index()

sns.barplot(x='category', y='rating', data=avg_rating)
plt.title("Average Rating by Category")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='brand', data=df, order=df['brand'].value_counts().index[:20])
plt.title("Product Count per Brand (Top 20)")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
corr = df.select_dtypes(include=['int64','float64']).corr()

sns.heatmap(corr, annot=True, linewidths=0.5, cmap="coolwarm")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```
